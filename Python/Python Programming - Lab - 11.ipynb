### 01) WAP to demonstrate the use of Pie chart.

```python
x=[10,20,30,40]
y=['Food','Travel','Acomodetion','Shopping']
c=['r','g','y','m']
ex=[0.5,0.0,0.0,0.0]
plt.pie(x,labels=y,colors=c,explode=ex,shadow=True,autopct="%0.2f%%",textprops={"fontsize":15,"color":"b"},labeldistance=1.5)
plt.title('Expences')
plt.show()
```

### 02) WAP to to Plot List random of X, Y Coordinates in Matplotlib.

```python
import random as rd
x=[10,20,30,40]
y=[rd.randrange(100) for i in range(0,4)]

plt.pie(x,labels=y)
plt.show()
```

### 03) WAP to demonstrate the use of Bar chart.

```python
x=["python","Java","C++","C"]
y=[80,70,60,82]
c=['r','g','y','m']

plt.bar(x,y,color=c,align='edge',edgecolor='b',alpha=0.5)
plt.title("Popularity of language")
plt.xlabel("Language")
plt.ylabel("Rating")
plt.show()
```

```python
x=["python","Java","C++","C"]
y=[80,70,60,82]
c=['r','g','y','m']

plt.barh(x,y,color=c,align='edge',edgecolor='b',alpha=0.5)
plt.title("Popularity of language")
plt.ylabel("Language")
plt.xlabel("Rating")
plt.show()
```

### 01) WAP to display the value of each bar in a bar chart using Matplotlib.

```python
x=["python","Java","C++","C"]
y=[80,70,60,82]
c=['r','g','y','m']

plt.bar(x,y,color=c,edgecolor='b',alpha=0.5)
def addlabels(x,y):
    for i in range(len(x)):
        plt.text(i,y[i],y[i])
plt.title("Popularity of language")
plt.xlabel("Language")
plt.ylabel("Rating")
addlabels(x,y)
plt.show()
```

### 02) WAP create a Scatter Plot with several colors in Matplotlib?

```python
x=[1,2,3,4,5]
y=[2,4,6,8,7]
c=['r','g','y','m','b']
l=[1,2,3,4,5]
plt.scatter(x,y,color=c,marker='*',linewidths=l,edgecolor='b')
plt.xlabel("X")
plt.ylabel("Y")
plt.show()
```
